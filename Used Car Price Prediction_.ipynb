# Model Building

Preprocessing New_Price

```python
def preprocess_new_price(value):
    if pd.notna(value):
        if 'Lakh' in value:
            return float(value.replace(' Lakh', '').replace(',', ''))  # Remove 'Lakh' and commas
        elif 'Cr' in value:
            return float(value.replace(' Cr', '').replace(',', '')) * 100  # Convert from Crores to Lakhs, remove 'Cr' and commas
    return None  # Handle missing (NaN) values

# Ensure 'New_Price' is treated as a string, apply the preprocessing function
df['New_Price'] = df['New_Price'].astype(str)
df['New_Price'] = df['New_Price'].apply(preprocess_new_price)
```

```python
#data where value of new price is given can be used as training data

cars=df[['Year', 'Kilometers_Driven', 'Fuel_Type',
       'Transmission', 'Owner_Type', 'Mileage', 'Engine', 'Power',
       'Price', 'Age', 'Brand','New_Price']]
```

```python
X_train = cars[cars['New_Price'].isnull()==False].drop(columns=['New_Price'])
y_train = cars[cars['New_Price'].isnull()==False]['New_Price']
```

```python
#data where value of new price is giving can be used to test model

df2=cars[cars['New_Price'].isnull()==True] 
```

```python
X_test=df2.drop(columns=['New_Price'])
```

```python
Fuel_Type = pd.get_dummies(X_train['Fuel_Type'],drop_first=True)
Transmission = pd.get_dummies(X_train['Transmission'],drop_first=True)

Owner_Type = pd.get_dummies(X_train['Owner_Type'],drop_first=True)
Brand = pd.get_dummies(X_train['Brand'],drop_first=True)
```

```python
X_train.drop(['Fuel_Type','Transmission','Owner_Type','Brand'],axis=1,inplace=True)

X_train = pd.concat([X_train,Fuel_Type,Transmission,Owner_Type,Brand],axis=1)
```

```python
Fuel_Type = pd.get_dummies(X_test['Fuel_Type'],drop_first=True)
Transmission = pd.get_dummies(X_test['Transmission'],drop_first=True)

Owner_Type = pd.get_dummies(X_test['Owner_Type'],drop_first=True)
Brand = pd.get_dummies(X_test['Brand'],drop_first=True)
```

```python
X_test.drop(['Fuel_Type','Transmission','Owner_Type','Brand'],axis=1,inplace=True)

X_test = pd.concat([X_test,Fuel_Type,Transmission,Owner_Type,Brand],axis=1)
```
